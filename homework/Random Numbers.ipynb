```python
def rng(N):
    values = []

    x = int(time() * 1e10)
    a = 7*7
    c = 0
    m = 2**31 - 1

    for i in range(N):
        x = (x*a + c) % m
        values.append(x)

    values = np.array(values)
    return values / m

N = 1000

values = 11 * rng(N) - 5
rounded_values = np.floor(values)

plt.figure(0)
plt.title("Before rounding")
plt.scatter(np.arange(N), values)
plt.figure(1)
plt.title("After rounding")
plt.scatter(np.arange(N), rounded_values)
plt.figure(2)
plt.title("Frequency of values")
plt.hist(values, 11)
```
