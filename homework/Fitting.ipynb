```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import numpy as np

f = lambda x: np.exp(-x/4) * np.cos(x)

# domain
x1 = 0
x2 = 10
num_of_points = 100
num_of_poly = 100
randomness = .6

# create points
x = np.linspace(x1, x2, num_of_points)
y = f(x) + (np.random.rand(num_of_points) - .5) * randomness

# ideal fit
y_ideal = f(x)

# polynomial fit
y_poly = []
for i in range(1, num_of_poly+1):
    y_poly.append(np.polyval(np.polyfit(x, y, i), x))

# plot
plt.figure(figsize=[8, 4])
plt.scatter(x, y, c='k', s=50)
plt.plot(x, y_ideal, label="Ideal")
plt.legend()
for i in range(len(y_poly)):
    plt.figure(figsize=[8, 4])
    plt.scatter(x, y, c='k', s=50)
    plt.plot(x, y_poly[i], label="Degree of " + str(i+1))
    plt.legend()
```

The ideal function to fit this data is a damped oscillation function. However, in terms of which degree polynomial does the best to fit the data, it appears that a degree of four does the best for this particular example. Polynomial's with degree numbers higher than four begin to have extra "wavy" artifacts resulting in overfitting.
