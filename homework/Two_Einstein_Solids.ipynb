```python
import matplotlib.pyplot as plt

def factorial(x):
    result = 1
    for i in range(1, x + 1):
        result *= i
    return result

def omega(N, q):
    return factorial(q + N - 1) / (factorial(q) * factorial(N - 1))

def plot_einstein(N_A, N_B, q_total, style="b-"):
    q_A = 0
    q_B = q_total

    q_A_list = []
    omega_total_list = []

    for i in range(q_total + 1):
        omega_A = omega(N_A, q_A)
        omega_B = omega(N_B, q_B)
        omega_total = omega_A * omega_B

        q_A_list.append(q_A)
        omega_total_list.append(omega_total)

        q_A += 1
        q_B -= 1

    plt.plot(q_A_list, omega_total_list, style)
    plt.xlabel("$q_A$")
    plt.ylabel("$\Omega(total)$")
```

```python
N_A = 300
N_B = 200
q_total = 100
plot_einstein(N_A, N_B, q_total)
```

```python
N_A = 3000
N_B = 2000
q_total = 100
plot_einstein(N_A, N_B, q_total, "r--")
```
