```python
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
```

```python
def z_prime(z, c):
    return z * z + c

def magnitude(z):
    return sqrt(z.real ** 2 + z.imag ** 2)

N = 1000
iterations = 20
x_min = -2
x_max = 1
y_min = -1.5
y_max = 1.5

# fill input space
input_space = np.zeros((N, N, 2))
x_values = np.linspace(x_min, x_max, N)
y_values = np.linspace(y_min, y_max, N)
for i in range(N):
    for j in range(N):
        input_space[i][j] = np.array([y_values[i], x_values[j]])

# fill output space
output_space = np.zeros((N, N))
for i in range(N):
    for j in range(N):
        z = complex(0, 0)
        c = complex(input_space[i][j][1], input_space[i][j][0])
        
        for k in range(iterations):
            z = z_prime(z, c)
        
        try:
            z_mag = magnitude(z)
        except:
            z_mag = 0
        if z_mag <= 2:
            output_space[i][j] = z_mag

plt.imshow(output_space, cmap="summer")
```
