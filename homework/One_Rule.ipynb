```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np

dataset = load_iris()
x = dataset.data
x = np.array( x>x.mean(), dtype='int')
prop_names = dataset.feature_names
y = dataset.target
data = {prop_names[0] : x[:,0], prop_names[1] : x[:,1], prop_names[2] : x[:,2], prop_names[3] : x[:,3], "Label" : y}
freq_dict = {}
for i in range(4):
    freq_dict[prop_names[i]] = np.array([np.array([0,0,0]), np.array([0,0,0])])
    for n in range(len(x[:,i])):
        target = y[n]
        freq_dict[prop_names[i]][x[n,i]][target] += 1
        
# create predictors
predictor_dict = {}
for i in range(4):
    max_list = []
    for j in range(2):
        max_list.append((freq_dict[prop_names[i]][j]).argmax())
    predictor_dict[prop_names[i]] = np.asarray(max_list)
    
# calculate errors
errors = []
for i in range(4):
    total = 0
    for x in range(2):
        for y in range(3):
            if predictor_dict[prop_names[i]][x] != y:
                total += freq_dict[prop_names[i]][x][y]
    errors.append(total)
errors = np.asarray(errors)

'''

[Below average, Above average]
0 = setosa
1 = versicolor
2 = virginica

'''

print("Predictors", predictor_dict, '\n')
print("Errors:", errors, '\n')
print("Best Predictor:", prop_names[errors.argmin()], "\t", predictor_dict[prop_names[errors.argmin()]])

df = pd.DataFrame(data)
df
```
